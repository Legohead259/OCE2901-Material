```python
# Part One: Imports and Constants
import numpy as np
import matplotlib.pyplot as plt
from math import pi

L = np.arange(150, 250, 10) # Wavelength [m]
k = 2*pi / L # Wave number
```

```python
# Part Two: Plots
h = 100 # MWL [m]
z = np.arange(0, -h-0.1, -0.1)

Kp = np.zeros([len(k), len(z)])

fig, ax_100 = plt.subplots(1, 1, figsize=(8, 6))
fig.suptitle("Pressure Response Factor as a Function of Depth")
ax_100.set_xlabel("Pressure response Factor (Kp)")
ax_100.set_ylabel("Water Depth (z)")
ax_100.invert_xaxis()
ax_100.grid()

fig2, ax_3 = plt.subplots(1, 1, figsize=(8, 6))
fig2.suptitle("Pressure Response Factor as a Function of Depth")
ax_3.set_xlabel("Pressure response Factor (Kp)")
ax_3.set_ylabel("Water Depth (z)")
ax_3.invert_xaxis()
ax_3.grid()

counter = 0
for k_i in k:
    Kp[counter] = np.divide(np.cosh(k_i*(h+z)), np.cosh(k_i*h))
    
    ax_100.plot(Kp[counter], z)
    ax_3.plot(Kp[counter, 0:30], z[0:30])

    counter += 1

ax_100.legend(["L = 150m", "L = 160m", "L = 170m", "L = 180m", "L = 190m", "L = 200m", "L = 210m", "L = 220m", "L = 230m", "L = 240m", "L = 250m"])
ax_3.legend(["L = 150m", "L = 160m", "L = 170m", "L = 180m", "L = 190m", "L = 200m", "L = 210m", "L = 220m", "L = 230m", "L = 240m", "L = 250m"])
```
