# Assignment 3: Dice Roll
## Part 1
For this first part, we will write a script that simulates rolling a dice and performs a basic statical analysis and generates a bar plot. We will roll the dice 10 times, then 1000 times and finish with an output table and figure that show all the relvant information

```python
# Import necessary libraries and set constants
import matplotlib.pyplot as plt
import numpy as np
from random import randint
from tabulate import tabulate
```

```python
def roll_die(itr: int=1) -> tuple:
    """
    A function that, when called, will simulate rolling a 6-sided die a given number of times. It will save every roll to a numpy array and return the array along with statistical parameters.

    ### Arguments:
    itr : int
        is the the number of times the dice should be rolled. Defaults to 1

    ### Returns:
    tuple object of the following
     - The ndarray of simulated die rolls
     - The statistical mean of the die rolls
     - The statistical standard deviation of the die rolls 
    """
    _arr = np.zeros(itr)
    for i in range(itr):
        _arr[i] = randint(1, 6)
    return (_arr, np.mean(_arr), np.std(_arr))

roll_10_arr, roll_10_mean, roll_10_stdev = roll_die(10)
roll_1000_arr, roll_1000_mean, roll_1000_stdev = roll_die(1000)

# Generate output table
table = [["10 Rolls", "1000 Rolls"], 
        ["Mean", roll_10_mean, roll_1000_mean],
        ["Stdev", roll_10_stdev, roll_1000_stdev]]

print(tabulate(table, headers='firstrow', tablefmt='fancy_grid'))

# Generate histogram figure
fig_one_die, (ax1, ax2) = plt.subplots(1,2, figsize=(8, 4))
fig_one_die.suptitle("Singular Dice Roll")
fig_one_die.supylabel("Occurrences")

labels, counts = np.unique(roll_10_arr, return_counts=True)
ax1.set_title("10 Rolls")
ax1.bar(labels, counts)
ax1.set_xticks(range(1,7,1))
ax1.set_yticks(range(0,max(counts)+1,1))

labels, counts = np.unique(roll_1000_arr, return_counts=True)
ax2.set_title("1000 Rolls")
ax2.bar(labels, counts)
ax2.set_xticks(range(1,7,1))
ax2.set_yticks(range(0,max(counts)+10,20))

plt.show()
plt.close(fig_one_die)
```

## Part 2
For this part, we will be taking two die, rolling them and then summing their values. Then, we will run the same statistical analysis as before and generate corresponding bar plots.

```python
def roll_two_dice_sum(itr: int=1) -> tuple:
    """
    A function that, when called, will simulate rolling two 6-sided die and summing their values, a given number of times. It will save every summation to a numpy array and return the array along with statistical parameters.

    ### Arguments:
    itr : int
        is the the number of times the dice should be rolled. Defaults to 1

    ### Returns:
    tuple object of the following
     - The ndarray of summed simulated die rolls
     - The statistical mean of the summed die rolls
     - The statistical standard deviation of the summed die rolls 
    """
    _arr = np.zeros(itr)
    for i in range(itr):
        _arr[i] = randint(1, 6) + randint(1,6)
    return (_arr, np.mean(_arr), np.std(_arr))

sum_10_arr, sum_10_mean, sum_10_stdev = roll_two_dice_sum(10)
sum_1000_arr, sum_1000_mean, sum_1000_stdev = roll_two_dice_sum(1000)

# Generate output table
table = [["10 Rolls", "1000 Rolls"], 
        ["Mean", sum_10_mean, sum_1000_mean],
        ["Stdev", sum_10_stdev, sum_1000_stdev]]

print(tabulate(table, headers='firstrow', tablefmt='fancy_grid'))

# Generate histogram figure
fig_one_die, (ax1, ax2) = plt.subplots(1,2, figsize=(8, 4))
fig_one_die.suptitle("Summed Dice Rolls")
fig_one_die.supylabel("Occurrences")

labels, counts = np.unique(sum_10_arr, return_counts=True)
ax1.set_title("10 Rolls")
ax1.bar(labels, counts)
ax1.set_xticks(range(1,13,1))
ax1.set_yticks(range(0,max(counts)+1,1))

labels, counts = np.unique(sum_1000_arr, return_counts=True)
ax2.set_title("1000 Rolls")
ax2.bar(labels, counts)
ax2.set_xticks(range(1,13,1))
ax2.set_yticks(range(0,max(counts)+10,20))

plt.show()
plt.close(fig_one_die)
```
